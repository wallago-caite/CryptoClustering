```python
#importing and fixing os due to error on MKL dataleak.  Currently just annoying but this fixes it.  
import os

# Set the environment variable to limit the number of threads to 1
os.environ["OMP_NUM_THREADS"] = "1"

# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler  = StandardScaler()
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaler.fit_transform(df_market_data))

df_scaled.head(3)
```

```python
# Copy the crypto names from the original data
crypto_names = df_market_data.index.tolist() # first column is the index, not a column so had to use index
crypto_names
```

```python
#find column names
dfscolumns = df_market_data.columns.tolist()
dfscolumns
```

```python
# Set the coinid column as index
df_scaled['coin_id'] = crypto_names

df_scaled.set_index('coin_id', inplace=True)

# set column names 
df_scaled.columns = dfscolumns


# Display sample data
df_scaled.head(5)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k: 
    model_kmeans = KMeans(n_clusters=i, random_state=1, n_init=10)
    model_kmeans.fit(df_scaled)
    inertia.append(model_kmeans.inertia_)

inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k , "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
O_ELBOW = elbow_df.hvplot.line(x = "k", y= "inertia", title = "elbow curve", xticks = k)

O_ELBOW
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value of "k" is 4.

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k

#function from TED class.... we break it down in this challenge
# def cluster_selection_graph(df, number: int, x_col: str, y_col: str):
#     model_kmeans = KMeans(n_clusters=number, random_state=1)
#     model_kmeans.fit(df)
#     y_pred_kmeans = model_kmeans.predict(df) # y is the target, pred is prediction, kmeans is the model
#     df["y_pred_kmeans"]= y_pred_kmeans 
#     graph = df.hvplot.scatter(x=x_col, y=y_col, by = "y_pred_kmeans")
#     return graph

model_kmeans = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model_kmeans.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
y_pred_kmeans = model_kmeans.predict(df_scaled)

# Print the resulting array of cluster values.
y_pred_kmeans
```

```python
# Create a copy of the DataFrame
df_cluster_prediction = df_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_cluster_prediction['predicted_clusters'] = y_pred_kmeans
```

```python
# Display sample data
df_cluster_prediction.head(3)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

blue_shades = ['#add8e6', '#87ceeb', '#4682b4', '#5f9ea0', '#0000ff']

O_PLOT = df_cluster_prediction.hvplot.scatter(
    title='Cluster Prediction: 24h vs 7d Price Change',
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='predicted_clusters',
    hover_cols='coin_id',
    color=blue_shades
)

O_PLOT
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)

pca
```

```python
# Use the PCA model with `fit_transform` to reduce to three principal components.
df_pca = pd.DataFrame(pca.fit_transform(df_cluster_prediction))

# View the first five rows of the DataFrame. 
df_pca.head(5)
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_

# pca1 explains 95% of the variation
# pca2 explains 1.8% of the variation
# pca3 explains 0.7% of the variation
```

```python
# Total explained variance 
total_explained_var = sum(pca.explained_variance_ratio_)

total_explained_var
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 88.442% (total_explained_var)

```python
# Create a new DataFrame with the PCA data.
df_pca2 = df_pca.copy()

# Copy the crypto names from the original data
# why would we do this? We already have the names-- we're going to call them 
crypto_names

# Set the coinid column as index
df_pca2['coin_id'] = crypto_names

df_pca2.set_index('coin_id', inplace=True)

# set column names 
df_pca2.columns = ['pca1','pca2','pca3']


# Display sample data
df_pca2.head(5)
# Display sample data
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia2 = []


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    k_model.fit(df_pca2)
    inertia2.append(k_model.inertia_)

```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data2 = {"k": k, "inertia": inertia2}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df2 = pd.DataFrame(elbow_data2)
elbow_df2.head(3)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
PCA_ELBOW = elbow_df2.hvplot.line(x = "k", y= "inertia", title = "elbow curve", xticks = k)

PCA_ELBOW
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value of 'k' using PCA data is 4.  


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It doesn't differ but it is MUCH clearer to determine.  

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_kmeans2 = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model_kmeans2.fit(df_pca2)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
y_pred_kmeans2 = model_kmeans2.predict(df_pca2)

# Print the resulting array of cluster values.
y_pred_kmeans2
```

```python
# Create a copy of the DataFrame with the PCA data
df_cluster_prediction_pca = df_pca2.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_cluster_prediction_pca['predicted_clusters'] = y_pred_kmeans2

# Display sample data
df_cluster_prediction_pca.head(3)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

red_shades = ['#ffcccc', '#ff9999', '#ff6666', '#ff3333', '#ff0000']

PCA_PLOT = df_cluster_prediction_pca.hvplot.scatter(
    title='Cluster Prediction: PCA1 vs PCA2',
    x="pca1",
    y="pca2",
    by='predicted_clusters',
    hover_cols='coin_id',
    color=red_shades
)

PCA_PLOT
```
